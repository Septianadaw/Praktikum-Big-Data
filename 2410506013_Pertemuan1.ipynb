```python
from pyspark.sql import SparkSession

# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame sederhana
data = [("Ali", 34), ("Budi", 23), ("Citra", 29), ("Dina", 45)]
columns = ["Nama", "Usia"]
df = spark.createDataFrame(data, columns)

# Menampilkan DataFrame
df.show()

# Filtering data
df_filtered = df.filter(df['Usia'] > 30)
df_filtered.show()

# Menghitung rata-rata usia
from pyspark.sql.functions import avg
df.groupBy().agg(avg("Usia")).show()

# Mengurutkan data berdasarkan usia
df_sorted = df.orderBy("Usia", ascending=False)
df_sorted.show()
```

```python
# Tugas 2
from pyspark.sql import SparkSession
from pyspark.sql.functions import avg

# Membuat Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame sederhana
data = [("Ali", 34), ("Budi", 23), ("Citra", 29), ("Dina", 45)]
columns = ["Nama", "Usia"]
df = spark.createDataFrame(data, columns)

print("Data Awal:")
df.show()

# Filtering (menyaring data usia > 30)
print("Data dengan usia lebih dari 30:")
df_filtered = df.filter(df['Usia'] > 30)
df_filtered.show()

# Menghitung rata-rata usia
print("Rata-rata usia:")
df.groupBy().agg(avg("Usia")).show()

# Mengurutkan data berdasarkan usia (descending)
print("Data diurutkan berdasarkan usia dari terbesar ke terkecil:")
df_sorted = df.orderBy("Usia", ascending=False)
df_sorted.show()
```

```python
# tugas 5
import pandas as pd

# Memulai Spark session
spark = SparkSession.builder.appName("GabunganPySparkPandas").getOrCreate()

# Membuat DataFrame PySpark
data_spark = [("Ali", 34), ("Budi", 23), ("Citra", 29), ("Dina", 45)]
columns_spark = ["Nama", "Usia"]
df_spark = spark.createDataFrame(data_spark, columns_spark)

# Membuat DataFrame Pandas
data_pandas = {"Nama": ["Ali", "Budi", "Citra", "Dina"],
               "Pekerjaan": ["Dokter", "Guru", "Insinyur", "Perawat"]}
df_pandas = pd.DataFrame(data_pandas)

# Konversi DataFrame Pandas ke PySpark
df_pandas_spark = spark.createDataFrame(df_pandas)

# Menggabungkan DataFrame PySpark dengan DataFrame hasil konversi Pandas
df_joined = df_spark.join(df_pandas_spark, on="Nama")

# Menampilkan hasil join
print("Hasil Gabungan Data:")
df_joined.show()

# Analisis sederhana menghitung rata-rata usia
print("Rata-rata Usia:")
df_joined.select(avg("Usia")).show()
```

```python
# Tugas 6
from pyspark.sql import SparkSession
import pandas as pd
from pyspark.sql.functions import max

# Membuat Spark session
spark = SparkSession.builder.appName("GabunganPySparkPandas").getOrCreate()

# DataFrame dari PySpark
data_spark = [("Ali", 34), ("Budi", 23), ("Citra", 29), ("Dina", 45)]
columns_spark = ["Nama", "Usia"]
df_spark = spark.createDataFrame(data_spark, columns_spark)

# DataFrame dari Pandas
data_pandas = {
    "Nama": ["Ali", "Budi", "Citra", "Dina"],
    "Pekerjaan": ["Dokter", "Guru", "Insinyur", "Perawat"]
}
df_pandas = pd.DataFrame(data_pandas)

# Konversi Pandas  PySpark
df_spark_from_pandas = spark.createDataFrame(df_pandas)

# Join berdasarkan kolom "Nama"
df_joined = df_spark.join(df_spark_from_pandas, on="Nama", how="inner")

# Menampilkan hasil join
print("Hasil Gabungan Data:")
df_joined.show()

# Operasi Statistik: Nilai Maksimum Usia
print("Nilai Maksimum Usia:")
df_joined.agg(max("Usia").alias("Maksimum Usia")).show()
```
