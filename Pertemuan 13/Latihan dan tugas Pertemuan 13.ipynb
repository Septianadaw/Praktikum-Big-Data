```python
# Import Library & Load Dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = '/home/septianadaw/Titanic-Dataset.csv'
df = pd.read_csv(file_path)
```

```python
# Multi-Bar Chart (Matplotlib)
# Perbandingan jumlah penumpang Male vs Female berdasarkan status Survived.

# Grouping data
grouped = df.groupby(['Sex', 'Survived']).size().unstack()

# Plot
plt.figure(figsize=(8, 6))
grouped.plot(kind='bar', figsize=(8, 6), color=['skyblue', 'salmon'])

plt.title('Perbandingan Jumlah Penumpang Berdasarkan Jenis Kelamin dan Status Survived')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Jumlah Penumpang')
plt.xticks(rotation=0)
plt.legend(['Tidak Selamat', 'Selamat'])
plt.grid(axis='y', linestyle='--', linewidth=0.5)

plt.show()
```

```python
# Scatter Plot (Seaborn)
# Relasi antara Age vs Fare dengan pewarnaan berdasarkan Survived.

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df, palette='coolwarm')

plt.title('Scatter Plot Age vs Fare Berdasarkan Status Survived')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True, linestyle='--', linewidth=0.5)

plt.show()
```

```python
# Custom Layout & Color Scheme

plt.figure(figsize=(10, 6))

sns.boxplot(x='Pclass', y='Fare', data=df, palette='Set2')

plt.title('Distribusi Fare Berdasarkan Pclass')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.grid(axis='y', linestyle='--', linewidth=0.5)

plt.show()
```

```python
# Membuat dashboard dengan beberapa visualisasi untuk meringkas dataset pilihan Anda.

# Grafik 1: Pie Chart (Distribusi Jenis Kelamin)
# Menampilkan persentase penumpang pria vs wanita.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/home/septianadaw/Titanic-Dataset.csv")

# Normalisasi nama kolom agar rapi
df.columns = df.columns.str.lower().str.replace(" ", "_")

plt.figure(figsize=(6, 6))
gender_counts = df['sex'].value_counts()

plt.pie(gender_counts,
        labels=['Laki-laki', 'Perempuan'],
        autopct='%1.1f%%',
        startangle=90)

plt.title("Persentase Jenis Kelamin Penumpang")
plt.show()
```

```python
# Grafik 2: Heatmap Korelasi
# Menampilkan hubungan antar variabel numerik di dataset Titanic.

plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Heatmap Korelasi Antar Variabel Numerik", fontsize=12)
plt.show()
```

```python
# Grafik 3: KDE Plot (Distribusi Umur)
# Menunjukkan bentuk distribusi umur penumpang dengan kurva density.

plt.figure(figsize=(7, 5))
sns.kdeplot(data=df, x="age", fill=True)
plt.title("Distribusi Umur Penumpang (KDE Plot)")
plt.xlabel("Umur")
plt.ylabel("Kepadatan")
plt.show()
```

```python
# Grafik 4: Violin Plot (Distribusi Usia Berdasarkan Kelas)
# Menampilkan distribusi usia per kelas

plt.figure(figsize=(7, 5))
sns.violinplot(data=df, x="pclass", y="age")
plt.title("Distribusi Usia Berdasarkan Kelas (Violin Plot)")
plt.xlabel("Kelas Penumpang")
plt.ylabel("Usia")
plt.show()
```
