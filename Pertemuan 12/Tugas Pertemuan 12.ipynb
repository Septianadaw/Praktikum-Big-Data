```python
# Tugas Latihan
# Memuat dataset pilihan Anda dan mengidentifikasi nilai yang hilang.

import pandas as pd

# Menentukan lokasi file dataset
file_path = '/home/septianadaw/Titanic-Dataset.csv'

# Memuat dataset
df = pd.read_csv(file_path)

# Menampilkan informasi awal dataset
print("\nInformasi Dataset: ")
print(df.info())

# Mengidentifikasi nilai yang hilang (missing values)
print("\nJumlah Nilai yang Hilang di Setiap Kolom: ")
print(df.isnull().sum())

# Menampilkan contoh beberapa baris data
print("\nContoh Beberapa Baris Data: ")
print(df.head())
```

```python
# Transformasi Data untuk Normalisasi Kolom Numerik

# Menyalin dataset agar data asli tidak berubah
df_normalized = df.copy()

# Menentukan kolom numerik
kolom_numerik = df_normalized.select_dtypes(include=['int64', 'float64']).columns

# Melakukan normalisasi dengan rumus: (x - min) / (max - min)
for kolom in kolom_numerik:
    nilai_min = df_normalized[kolom].min()
    nilai_max = df_normalized[kolom].max()
    df_normalized[kolom] = (df_normalized[kolom] - nilai_min) / (nilai_max - nilai_min)

# Menampilkan hasil normalisasi
print("\nDataset Setelah Normalisasi Kolom Numerik: ")
print(df_normalized.head())
```

```python
# Standarisasi Kolom Kategoris dan Menghapus Duplikat

# Menyalin dataset hasil normalisasi
df_bersih = df_normalized.copy()

# Menampilkan jumlah data awal sebelum pembersihan
print(f"Jumlah data awal sebelum pembersihan: {len(df_bersih)} baris")

# Menstandarkan kolom kategoris
kolom_kategoris = df_bersih.select_dtypes(include='object').columns
for kolom in kolom_kategoris:
    df_bersih[kolom] = df_bersih[kolom].str.capitalize()

# Menghapus data duplikat
df_bersih.drop_duplicates(inplace=True)

# Menampilkan hasil akhir
print("\nDataset Setelah Standarisasi dan Menghapus Duplikat: ")
print(df_bersih.head())
print(f"\nJumlah data akhir setelah pembersihan: {len(df_bersih)} baris")
```
